```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import statsmodels.api as sm

print("pandas version:", pd.__version__)
print("numpy version:", np.__version__)
print("All libraries loaded successfully!")
```

```python
# Sanity check chart — dummy Toronto price data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
toronto_prices = [1200, 1250, 1300, 1280, 1260, 1240]  # dummy data, in thousands CAD

plt.figure(figsize=(8, 4))
plt.plot(months, toronto_prices, marker='o', color='steelblue', linewidth=2)
plt.title('Test Chart — Toronto Average Price (dummy data)')
plt.ylabel('Price (CAD thousands)')
plt.grid(alpha=0.3)
plt.show()
```

```python
# Professional research-style chart setup
import matplotlib.pyplot as plt
import matplotlib as mpl

# Set global style for all charts in this project
mpl.rcParams['font.family'] = 'Helvetica'
mpl.rcParams['font.size'] = 11
mpl.rcParams['axes.titlesize'] = 13
mpl.rcParams['axes.titleweight'] = 'bold'
mpl.rcParams['axes.spines.top'] = False
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.grid'] = True
mpl.rcParams['grid.alpha'] = 0.25
mpl.rcParams['grid.linestyle'] = '--'

# Custom palette: one color per city, used consistently across the whole project
CITY_COLORS = {
    'Toronto':   '#C8102E',  # red
    'Vancouver': '#1B5E20',  # forest green
    'London':    '#1D3557',  # navy blue
    'New York':  '#E8A317',  # amber
}

# Dummy data just to see the style
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
toronto   = [1200, 1250, 1300, 1280, 1260, 1240, 1220, 1210, 1205, 1200, 1195, 1190]
vancouver = [1100, 1140, 1180, 1170, 1150, 1130, 1110, 1100, 1095, 1090, 1085, 1080]
london    = [980,  1010, 1050, 1040, 1020, 1000, 985,  975,  970,  965,  960,  955 ]
newyork   = [1300, 1330, 1370, 1355, 1340, 1320, 1305, 1295, 1290, 1285, 1280, 1275]

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(months, toronto,   label='Toronto',   color=CITY_COLORS['Toronto'],   linewidth=2, marker='o', markersize=4)
ax.plot(months, vancouver, label='Vancouver', color=CITY_COLORS['Vancouver'], linewidth=2, marker='o', markersize=4)
ax.plot(months, london,    label='London',    color=CITY_COLORS['London'],    linewidth=2, marker='o', markersize=4)
ax.plot(months, newyork,   label='New York',  color=CITY_COLORS['New York'],  linewidth=2, marker='o', markersize=4)

ax.set_title('Average Residential Price by City, 2024 (dummy data)', loc='left', pad=12)
ax.set_ylabel('Price (thousands USD)')
ax.set_xlabel('')
ax.legend(loc='upper right', frameon=False, ncol=4)

plt.tight_layout()
plt.show()
```
