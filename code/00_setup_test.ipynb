```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import statsmodels.api as sm

print("pandas version:", pd.__version__)
print("numpy version:", np.__version__)
print("All libraries loaded successfully!")
```

```python
# Sanity check chart — dummy Toronto price data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
toronto_prices = [1200, 1250, 1300, 1280, 1260, 1240]  # dummy data, in thousands CAD

plt.figure(figsize=(8, 4))
plt.plot(months, toronto_prices, marker='o', color='steelblue', linewidth=2)
plt.title('Test Chart — Toronto Average Price (dummy data)')
plt.ylabel('Price (CAD thousands)')
plt.grid(alpha=0.3)
plt.show()
```
